# Get collection by name

```python
def find_collection_by_tag(history_id: str, tag: str):
    """
    Find a dataset collection in a history whose tags include the desired tag.
    """
    # List all history content including dataset collections
    contents = gi.histories.show_history(history_id, contents=True, types=['dataset_collection'])

    tag = f"name:{tag}"

    for item in contents:
        # Some objects include a "tags" field (Galaxy may return it in content)
        item_tags = item.get('tags') or []
        if tag in item_tags:
            return item['id'], item

    return None, None

collection_id, collection_ = find_collection_by_tag(history_id, tag)
print(collection_id)
```
